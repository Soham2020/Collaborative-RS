```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("ggplot")

import sklearn
from sklearn.decomposition import TruncatedSVD
from google.colab import files
uploaded=files.upload()
S_Mart_ratings = pd.read_csv('data.csv',encoding='1252')
S_Mart_ratings = S_Mart_ratings.dropna()
S_Mart_ratings.head()
```

```
S_Mart_ratings.shape
```

```
S_Mart_ratings_new = S_Mart_ratings.head(10000)
```

```
#Utility Matrix
ratings_utility_matrix = S_Mart_ratings_new.pivot_table(values='rating', index='userId', columns='productId', fill_value=0)
ratings_utility_matrix.head()
```

```
ratings_utility_matrix.shape
```

```
#Transposing
temp = ratings_utility_matrix.T
temp.head()
```

```
#Decomposing of Matrix
SVD = TruncatedSVD(n_components=8)
decomposed_matrix = SVD.fit_transform(temp)
decomposed_matrix.shape
```

```
#Correlation Matrix
correlation_matrix = np.corrcoef(decomposed_matrix)
correlation_matrix.shape
```

```
# Randomly selection product id '61dad0f2bd0376ae20bb8261' from correlation matrix
temp.index[3]
```

```
chosen = temp.index[3]

product_names = list(temp.index)
product_ID = product_names.index(chosen)
product_ID
```

```
correlation_product_ID = correlation_matrix[product_ID]
correlation_product_ID.shape
```

```
# Recommending top 2 highly correlated products for product '61dad0f2bd0376ae20bb8225'
Recommend = list(temp.index[correlation_product_ID > 0.40])

Recommend.remove(chosen) 

S_Mart_ratings_new.loc[S_Mart_ratings_new['productId']==Recommend[1]][['title','price']]
```

```
S_Mart_ratings_new.loc[S_Mart_ratings_new['productId']==Recommend[0]][['title','price']]
```
